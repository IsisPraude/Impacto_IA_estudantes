```python
import pandas as pd

df = pd.read_csv('AI_Impact_Student_Life_2026.csv')
df['Difference_GPA'] = df['GPA_Baseline'] - df['GPA_Post_AI']
df_notas = df.drop(columns=['Student_ID'])
display(df_notas)
display(df_notas.info())
```

```python
# Filtragem de dados para relacionar confiança na carreira, GPA, frequência de uso e idade

df_confidence = df_notas[['Age','Major','Task_Frequency_Daily','GPA_Baseline','Career_Confidence_Score']]
df_confidence.to_csv('df_confidence.csv')
display(df_confidence)
```

```python
# Filtragem de dados para análise de relação entre curso, frequência de uso e tempo salvo

df_efficiency = df_notas[['Age','Major','Task_Frequency_Daily','Time_Saved_Hours_Weekly','Difference_GPA']]
df_cursos = df_efficiency[['Major','Time_Saved_Hours_Weekly','Task_Frequency_Daily']].groupby(['Major','Task_Frequency_Daily']).mean()
df_cursos.to_csv('df_cursos.csv')
display(df_efficiency)
display(df_cursos)
display(df_cursos.describe())
```

```python
# Filtragem de dados para relacionar para quais tarefas e quais ferramentas de IA mais são utilizadas por cada curso

df_uso = df_notas[['Major','Primary_AI_Tool','Main_Usage_Case']].groupby(['Major','Primary_AI_Tool','Main_Usage_Case']).size().reset_index()
display(df_uso)
df_uso.to_csv('df_uso.csv')
```
